```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
anneal = np.load("result_anneal.npz")
no_anneal = np.load("result_non_anneal.npz")
```

```python
plt.plot(np.power(10,anneal["logbetas"]),anneal["result_K"], label="anneal K")
plt.plot(np.power(10,anneal["logbetas"]),anneal["result_D"], label="anneal D")
plt.plot(np.power(10,no_anneal["logbetas"]),no_anneal["result_K"], label="non-anneal K")
plt.plot(np.power(10,no_anneal["logbetas"]),no_anneal["result_D"], label="non-anneal D")
plt.ylim([0,80])
plt.semilogx()
plt.legend()
plt.show()
```
