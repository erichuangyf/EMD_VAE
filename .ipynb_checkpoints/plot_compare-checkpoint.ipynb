```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
anneal_result=np.load("result_anneal.npz")
non_anneal_result= np.load("result_non_anneal.npz")
```

```python
logbetas = np.linspace(1,-6,50).astype(np.float32)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.power(10,logbetas),anneal_result["result_D"], label="anneal")
plt.plot(np.power(10,logbetas),non_anneal_result["result_D"], label="non-anneal")
plt.legend()
plt.semilogx()
plt.title("heat capacity D")
plt.ylim([0,None])
plt.savefig(os.path.join("plots", "result_d.png"))
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.power(10,logbetas),anneal_result["result_K"], label="anneal")
plt.plot(np.power(10,logbetas),non_anneal_result["result_K"], label="non-anneal")
plt.legend()
plt.semilogx()
plt.savefig(os.path.join("plots", "result_k.png"))
plt.title("heat capacity K")
plt.ylim([0,None])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.power(10,logbetas),anneal_result["kl_loss"], label="anneal")
plt.plot(np.power(10,logbetas),non_anneal_result["kl_loss"], label="non-anneal")
plt.legend()
plt.semilogx()
plt.title("KL loss")
plt.ylim([0,None])
plt.savefig(os.path.join("plots", "kl_loss.png"))
plt.show()
```
