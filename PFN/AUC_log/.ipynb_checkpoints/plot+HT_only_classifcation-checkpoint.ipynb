```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
aucs = np.loadtxt("auc_log.csv")
```

```python
logbetas = np.linspace(1,-6,30).astype(np.float32)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.power(10,logbetas),1-aucs)
plt.semilogx()
plt.ylabel("$\log(1-AUC)$" ,fontsize=40)
plt.xlabel(r"$\beta$" ,fontsize=40)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.yscale("log")
plt.savefig(os.path.join("1-auc_beta.pdf"), dpi=200, bbox_inches="tight")
plt.show()
```
